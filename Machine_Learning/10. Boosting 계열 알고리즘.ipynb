# 랜덤 포레스트
- 정형 데이터를 다루는 대표적인 머신러닝 알고리즘
- 결정 트리를 무작위로 만들어 숲을 만드는 알고리즘
- CART 계열: RandomForestClassifier, RandomForestRegressor
- 부트스트랩 샘플로 훈련을 진행한다.
- 부트스트랩 샘플이란 훈련 세트에서 중복을 허용해 무작위로 샘플링한 데이터 샘플을 뜻한다.
- 각각의 부트스트랩 샘플들로 훈련한 모델들의 예측 정확도를 모두 더해 트리의 개수로 나누면 평균 정확도가 산출된다.
- 특성 개수의 루트만큼 무작위로 골라 최선의 특성 분할 방법을 찾아서 훈련한다.
- 모든 노드를 만들 때마다 이 과정을 반복해 최선의 특성 분할을 적용하는 알고리즘

```python
wine = pd.read_csv('https://raw.githubusercontent.com/rickiepark/hg-mldl/master/wine.csv')
wine.head()
```

```python
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
train_input.shape, test_target.shape
```

```python
train_target.shape, test_target.shape
```

**분류 알고리즘을 사용해 분류할 때 그 찾아낼 대상을 양성클래스라고 부르고, 1의 정형값을 갖는다.**

# 랜덤포레스트 분류기를 이용한 화이트와인 분류
- 화이트와인: 1

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트는 기본적으로 100개의 결정트리를 사용함(속도 느림)
# 그래서 내가 가지고 있는 모든 CPU 코어를 사용하겠다는 n_jobs=-1 옵션을 줌
rf = RandomForestClassifier(n_jobs=-1, random_state=42)

# return_train_score=True: 검증세트 뿐만아니라 훈련세트의 점수도 같이 반환한다.
scores = cross_validate(rf, train_input, train_target, n_jobs=-1, return_train_score=True)

# 검증세트와 훈련세트의 정확도를 통해 과대적합의 판별이 용이
print(np.mean(scores['train_score']), np.mean(scores['test_score'])) # 과대적합
```

```python
rf.fit(train_input, train_target)

# 훈련 시 피처별 중요도
print(rf.feature_importances_)
```

**피처들이 골고루 중요하게 사용됐다. 그 중 가장 중요하게 사용된 특성은 sugar(당도)이다.**

## 과대적합 해결법
1. 가지치기
- max_depth
- min_samples_split

2. gini 불순도

```python
# oob: 부트스트랩 샘플을 통해 샘플링한 데이터 중에서도 전혀 훈련에 사용되지 않은 데이터
## oob_score: oob를 통한 모델 자체 평가 점수
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

# 엑스트라 트리(Extra Tree)

- 랜덤포레스트와 비슷한 모델
- 랜덤포레스트와 거의 모든 과정이 비슷하지만 과대적합을 억제하는 방식에 차이가 있다.
- 랜덤포레스트처럼 부트스트랩 샘플을 사용하지 않고 전체 샘플을 사용한다.
- 정보 이득을 계산하지 않고, 여러 특성을 기준으로 랜덤하게 트리를 분할해 이중 최적의 불순도를 가지는 트리를 찾아내는 방법
- 그래서 랜덤포레스트에 비해 속도가 빠르다.
- ExtraTreesClassifier 클래스를 사용

```python
from sklearn.ensemble import ExtraTreesClassifier

# 랜덤포레스트와 마찬가지로 100개의 결정트리를 사용함
et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
et.fit(train_input, train_target)
print(et.feature_importances_)
```

- 엑스트라트리가 무작위성이 더 크기 때문에 랜덤포레스트보다 더 많은 결정트리를 훈련해야한다.
- 특성의 개수도 더 많이 주어질 때 랜덤포레스트 보다는 유리한 성능을 제공한다.

# 그레이디언트 부스팅(Gradient Boosting)
- 기본적으로 깊이가 3인 결정 트리 100개를 사용
- 깊이가 얕은 결정 트리를 사용하기 떄문에 과대적합에 강하고 일반적으로 높은 일반화 성능을 기대
- 분류에서는 로지스틱 손실함수를, 회귀에서는 평균제곱오차를 사용한다.
- 이전 트리의 손실 값들이 낮아지는 방향으로 회귀 결정 트리를 계속 추가하는 알고리즘
- GradientBoostingClassifier/Regressor 클래스
- max_depth 또는 learning_rate 파라미터를 추가해 과대적합을 억제할 수 있다.
- 학습률을 증가시키고 트리의 개수를 늘리면 좀 더 성능이 향상될 수 있다.

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 과소적합 해결을 위해 트리의 개수를 100에서 500으로 늘리고 학습률을 0.1에서 0.2로 올림
## n_estimators = 트리의 개수, default: 100
## learning_rate = 학습율, default: 0.1
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

# 히스토그램 기반 그레이디언트 부스팅(HGBM)
- Histogram-based Gradient Boosting: 정형 데이터를 다루는 머신러닝 알고리즘에서 가장 인기가 높은 알고리즘
- 훈련 데이터를 256개의 구간으로 나눠 학습시키는 알고리즘
- 트리의 개수를 지정하는 n_estimators 파라미터 대신 부스팅 횟수를 지정하는 max_iter를 사용
- 성능을 향상시키기 위해 max_iter 파라미터를 조정한다.

```python
import sys
# 파이썬 버전을 확인할 때
print('Python:', sys.version)
print('Pandas:', pd.__version__)
print('Numpy:', np.__version__)

import matplotlib as mpl
print('Matplotlib:', mpl.__version__)

import scipy as sp
print('Scipy:', sp.__version__)

import sklearn
print('Scikit-learn:', sklearn.__version__)
```

```python
# 사이킷런 1.0 버전 아래에서는
# from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

**HGBM은 feature_importances_라는 속성이 없다**

```python
# feature의 중요도를 별도의 함수를 이용하여 계산을 수행한다.
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)

# n_repeats: 랜덤하게 섞을 횟수
# permutation_importance: 반복해서 얻은 특성 중요도(importances)
result = permutation_importance(hgb, train_input, train_target,
                                n_repeats=10, random_state=42, n_jobs=-1)

# importances_mean: 중요도의 평균
# importances_std: 중요도의 표준편차
print(result.importances_mean)
```

```python
result = permutation_importance(hgb, test_input, test_target,
                                n_repeats=10, random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
hgb.score(test_input, test_target)
```

# XGBoost vs LightGBM
- 히스토그램 그레이디언트 부스팅 전용 라이브러리가 LightGBM, XGBosst, NGBoost...
- 트리 기반의 알고리즘의 앙상블 학습에서 각광받는 알고리즘 중 하나
- GBM에 기반하고 있지만, GBM의 단점인 느린 수행시간, 과적합 규제 등을 해결한 알고리즘
**주요장점**
- 뛰어난 예측 성능
- GBM 대비 빠른 수행시간
- 과적합 규제
- Tree pruning(트리 가지치기): 긍정 이득이 없는 분할을 가지치기 해서 분할 수 줄임
- 자체 내장된 교차검증
  - 반복 수행시마다 내부적으로 교차검증을 수행해 최적화된 반복 수행횟수를 가질 수 있다.
  - 지정된 반복횟수가 아니라 교차검증을 통해 평가 데이터세트의 평가 값이 최적화되면 반복을 중간에 멈출 수 있는 기능이 있다.
- 결손값 자체 처리
- XGBoost는 독자적인 XGBoost 모듈과 사이킷런 프레임워크 기반의 모듈이 존재한다.
- 독자적인 모듈은 고유의 API와 하이퍼파라미터를 사용하지만 사이킷런 기반 모듈에서는 다른 Estimator와 동일한 사용법을 가진다.
