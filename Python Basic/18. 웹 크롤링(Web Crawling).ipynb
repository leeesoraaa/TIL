## 모듈을 로딩

```python
from urllib.request import urlopen # 원하는 사이트에 접속할 목적으로 사용
from bs4 import BeautifulSoup # 사이트 접속한 페이지의 모든 값을 가져올 목적으로 사용
```

```python
naver = urlopen("https://www.naver.com")
print(naver.read())  # html 형식으로 읽어옴
```

```python
# 원하는 사이트에 접속해 모든 값을 가져오기
response = urlopen("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%AF%B8%EA%B5%AD+%EA%B0%9C%EC%9D%B8%EC%86%8C%EB%B9%84%EC%A7%80%EC%B6%9C%EB%AC%BC%EA%B0%80")
```

```python
# html 태그에 접근
soup = BeautifulSoup(response, "html.parser")
```

```python
value = soup.find("strong", {"class":"num"})
print(value)
```

```python
# 원하는 값만 가져오기
result = value.text.strip()  # strip(): 좌우 공백 제거 (= 코드도 함께 제거하고 값만 가져옴)
print(result) 
```

```python
print(result[:4]) # 리스트 형태임
```
