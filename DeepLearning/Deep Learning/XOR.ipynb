```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = np.array([[0,0],[1,0],[0,1],[1,1]])
y = np.array([[0],[1],[1],[0]])  # XOR
```

```python
model = Sequential()  # Sequential 객체 선언
model.add(Dense(1, input_dim=2, activation='sigmoid')) # 신경망 구조 생성
model.summary()  # 신경망 구조 보여줌
```

```python
model.compile(optimizer='adam',   # 오차를 줄여나갈 방법(경사하강법)
              loss='binary_crossentropy',  # 손실함수(교차엔트로피)
              metrics=['acc'])   # 평가지표
```

```python
model.fit(X, y, epochs=1000)
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.predict([[1,1]])
```

# 은닉층과 오차 역전파

## relu 활성화 함수

```python
model = Sequential()  # Sequential 객체 선언
model.add(Dense(32, input_dim=2, activation='relu')) # 은닉층
model.add(Dense(1, input_dim=2, activation='sigmoid')) # 신경망 구조 생성
model.summary()  # 신경망 구조 보여줌
```

```python
model.compile(optimizer='adam',   # 오차를 줄여나갈 방법(경사하강법)
              loss='binary_crossentropy',  # 손실함수(교차엔트로피)
              metrics=['acc'])
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.predict([[1,1]])
```

```python
model = Sequential()  # Sequential 객체 선언
model.add(Dense(32, input_dim=2, activation='relu')) # 은닉층(노드 32개)
model.add(Dense(12, input_dim=2, activation='relu')) # 은닉층(노드 12개)
model.add(Dense(1, input_dim=2, activation='sigmoid')) # 신경망 구조 생성
model.summary()  # 신경망 구조 보여줌
```

```python
model.compile(optimizer='adam',   # 오차를 줄여나갈 방법(경사하강법)
              loss='binary_crossentropy',  # 손실함수(교차엔트로피)
              metrics=['acc'])
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.predict([[1,1]])
```

## 하이퍼볼릭 탄젠트(tanh) 활성화 함수

```python
model = Sequential()  # Sequential 객체 선언
model.add(Dense(32, input_dim=2, activation='tanh')) # 은닉층(노드 32개)
model.add(Dense(12, input_dim=2, activation='tanh')) # 은닉층(노드 12개)
model.add(Dense(1, input_dim=2, activation='sigmoid')) # 신경망 구조 생성
model.summary()  # 신경망 구조 보여줌

model.compile(optimizer='adam',   # 오차를 줄여나갈 방법(경사하강법)
              loss='binary_crossentropy',  # 손실함수(교차엔트로피)
              metrics=['acc'])

model.fit(X, y, epochs=1000)
```

## 소프트플러스(softplus) 활성화 함수

```python
model = Sequential()  # Sequential 객체 선언
model.add(Dense(32, input_dim=2, activation='softplus')) # 은닉층(노드 32개)
model.add(Dense(12, input_dim=2, activation='softplus')) # 은닉층(노드 12개)
model.add(Dense(1, input_dim=2, activation='sigmoid')) # 신경망 구조 생성
model.summary()  # 신경망 구조 보여줌

model.compile(optimizer='adam',   # 오차를 줄여나갈 방법(경사하강법)
              loss='binary_crossentropy',  # 손실함수(교차엔트로피)
              metrics=['acc'])

model.fit(X, y, epochs=1000)
```
