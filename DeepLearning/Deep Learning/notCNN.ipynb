```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import os
import cv2
import numpy as np
```

```python
image_path = "/content/sample_data/image/"
csv_path = "/content/pokemon.csv"
```

```python
types = pd.read_csv(csv_path)

types.sort_values(by = "Name", inplace = True)
types.reset_index(inplace = True, drop = True)
types.Type2.fillna("None", inplace = True)
types.head()
```

```python
dir_imgs = os.listdir(image_path)
dir_imgs.sort()

fig, ax = plt.subplots(2, 5, figsize = (22, 10))
for i in range(0, 10):
    img = plt.imread(os.path.join(image_path, dir_imgs[i]))
    ax[i//5, i%5].imshow(img)
    ax[i//5, i%5].axis('off')
    ax[i//5, i%5].set_title("Name: %s\nType1: %s\nType2: %s" %(types.loc[i, "Name"].title(),
                                                              types.loc[i, "Type1"].title(),
                                                              types.loc[i, "Type2"].title()))
```

```python
dir_imgs = [os.path.join(image_path, path) for path in dir_imgs]
```

```python
def process_image(filepath):
    image = cv2.imread(filepath)
    return image

X = list(map(process_image, dir_imgs))
```
