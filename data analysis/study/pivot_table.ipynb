```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 실습데이터
df = pd.read_excel('sales-funnel.xlsx')
df.head()
```

- 피봇테이블(pivot table)이란 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것을 말한다.
- 판다스는 피봇테이블을 만들기 위한 pivot

```python
# 다중인덱스를 지정할 수도 있다.
pd.pivot_table(df, index=['Name', 'Manager', 'Rep'])
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'])
```

```python
# 특정 value만 지정해서 나타나도록 할 수도 있다.
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'])
```

- value를 pivot_table로 합친 경우 평균값이 기본이 된다.
- 합계를 사용하려면 aggfunc 옵션을 사용해서 np.sum

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], aggfunc=[np.mean, len])
```

```python
# index를 지정하고, values를 지정한다.
# 합과 평균을 표시하도록 지정하고 이 과정에서 빈칸이 나타나면 NaN으로 두지 말고 fill_value 옵션을 이용하여 0으로 채운다ㅓ.
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'],
               columns=['Product'], aggfunc=[np.sum])
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'],
               columns=['Product'], aggfunc=[np.sum], fill_value=0)
```

```python
pd.pivot_table(df, index=['Manager', 'Rep', 'Product'],
               values=['Price', 'Quantity'],
               aggfunc=[np.sum, np.mean], fill_value=0, margins=True)
```
