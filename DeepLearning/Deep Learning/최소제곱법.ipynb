# 공부 시간을 통한 점수 예측
- 최소제곱법

```python
x = np.array([2, 4, 6, 8]) # 공부한 시간
y = np.array([81, 93, 91, 97]) # 점수
```

```python
import numpy as np

# y = ax + b
# x와 y의 평균 계산
x_mean = np.mean(x)
y_mean = np.mean(y)

# 최소제곱법을 사용하여 기울기(a)와 y 절편(b) 계산
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
a = numerator / denominator
b = y_mean - (a * x_mean)
```

- 기울기 = x와 y의 공분산 / x의 분산
- numerator = x의 편차 * y의 편차 들의 합
- denominator = x의 편차 제곱 합

```python
print(a, b)
```

```python
def solution(x):
  y = a * x + b
  return y
```
