```python
from sklearn.datasets import load_iris

iris = load_iris()
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris_df = pd.DataFrame(iris.data, columns = cols)
iris_df.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, random_state = 0)

kmeans.fit(iris_df)
```

```python
kmeans.labels_
```

```python
iris_df['target'] = iris.target
iris_df['cluster'] = kmeans.labels_
iris_df.head()
```

```python
iris_result = iris_df.groupby(['target', 'cluster']).size()
iris_result
```

```python
# 차원축소
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

pca_transformed = pca.fit_transform(iris.data)

# 차원축소된 데이터 데이터프레임에 추가
iris_df['pca_x'] = pca_transformed[:, 0]
iris_df['pca_y'] = pca_transformed[:, 1]
iris_df.head()
```

```python
import seaborn as sns

sns.scatterplot(x = 'pca_x', y = 'pca_y', data = iris_df, hue = 'cluster',
                 style = 'cluster')
plt.show()

```

```python
pred = kmeans.predict(iris_df[['sepal_length', 'sepal_width', 'petal_lentgh', 'petal_width']])
pred
```

```python
# 군집 결과 시각화
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_length'], c = pred, cmap = 'summer')
plt.show()
```

```python
# 군집별 중앙값 추출
centers = kmeans.cluster_centers_
centers
```

```python
# 군집별 중앙값 시각화
plt.scatter(x = centers[:, 0], y = centers[:, 2], marker = '^', c = 'r')
plt.show()
```

## 1-1. 데이터프레임으로 변형

```python
data = wine.data
label = wine.target   # 분류에서는 target 대신 label 이라는 표현 씀
columns = wine.feature_names

df = pd.DataFrame(data, columns = columns)

print(df.shape)
```

## 1-2. 적잘한 k값을 찾는 방법 - 덴도그램 시각화

```python
from scipy.cluster.hierarchy import linkage, dendrogram

clusters_single = linkage(y = data, method = 'single', metric = 'euclidean')
clusters_complete = linkage(y = data, method = 'complete', metric = 'euclidean')
clusters_average = linkage(y = data, method = 'average', metric = 'euclidean')
```

```python
fig, axes = plt.subplots(3, 1)
fig.set_size_inches(15, 15) # 도화지 사이즈를 픽셀(기본)에서 인치로 변경

ax1 = dendrogram(clusters_single, ax = axes[0], leaf_font_size = 0)
ax2 = dendrogram(clusters_complete, ax = axes[1], leaf_font_size = 0)
ax3 = dendrogram(clusters_average, ax = axes[2], leaf_font_size = 0)
```

- wine 데이터셋은 single linkage는 피해야 한다.
- 군집 수는 3개 또는 4개가 적당해 보인다.

- n_components의 매개변수의 값에 따라 서로 다른 두가지 동작을 수행한다.
- 매개변수의 값이 1보다 크면 해당 개수만큼 특성이 반환된다.
- 매개변수의 값이 1보다 작으면 변동비율을 담을 수 있다. ex) 95% 보존 => n_comnponents = 0.95

## 1-6. 최적의 군집의 수(k) 찾기
- K-Means 알고리즘의 단점 중 하나는 클러스터의 개수를 사전에 지정해야 한다는 점이다.
- 군집알고리즘에서 적절한 k를 찾기 위한 완벽한 방법은 없다.
- 대표적인 방법 중 하나가 엘보우 방법
- 덴도그램은 보는사람의 관점에 따라 달라질 수 있어 k를 찾는 적절한 방법은 아니다.

**Elbow Method**
- K-Mneas 알고리즘은 클러스터 중심과 클러스터에 속한 샘플 사이의 거리를 잴 수 있다.
- 이 거리의 제곱의 합을 inertia라고 부른다.
- inertia(이너셔) : 클러스터에 속한 샘플이 얼마나 가깝게 모여 있는지를 나타내는 값
- 클러스터의 개수가 늘어나면 클러스터 개개의 거리의 크기는 줄어든다.(이너셔도 작아진다.)
- 엘보우 방법은 클러스터의 개수를 늘려가면서 이너셔의 변화를 관찰하고 최적의 클러스터의 수를 찾는다.
  - 클러스터의 개수를 증가시키면서 이너셔를 시각화하면 감소하는 속도가 꺾이는 지점이 있다.
  - 이 지점부터 클러스터 개수를 늘려도 클러스터에 잘 밀집된 정도가 크게 변화되지 않는다.
  - 즉 이너셔가 크게 줄어들지 않는다.
  - 이 지점이 팔꿈치 모양과 같아서 엘보우 방법이라고 부른다.

**용어정리**
- 클러스터(Cluster) : 비슷한 특성을 가진 데이터들끼리 묶는 것
- 클러스터링(Clustering) : 데이터들을 군집으로 묶어주는 작업
- 중심점, 도심(Centroid) : 클러스터의 중심

```python
# 엘보우 방법
inertia = []

for k in range(2, 16):
  km = KMeans(n_clusters = k, random_state = 200)
  km.fit(data)
  iner = km.inertia_
  inertia.append(iner)

  print('군집 수 : ', k, ' ', '이너셔 : ', iner)
```

```python
inertia = np.array(inertia)
```

```python
# 시각화
plt.plot(range(2, 16), inertia)

plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
