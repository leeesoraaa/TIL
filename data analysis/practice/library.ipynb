### 결측 데이터를 처리하는 방법
1. 결측된 데이터가 너무 많은 경우 해당 열 전체 삭제
2. 결측된데이터가 일부인 경우 가장 그럴듯한 값으로 대체

```python
# 데이터가 절반 이상이 없는 열 삭제
# thresh 옵션을 사용하면 특정 갯수 이상의 비결측 데이터가 있는 행 또는 열만 남길 때 사용
titanic = titanic.dropna(thresh=int(len(titanic) * 0.5), axis=1)
msno.matrix(titanic)
plt.show()
```

#### duplicated() 메서드는 기본적으로 모든 열을 기준으로 중복된 행을 찾는 메서드이다.
- subset이라는 옵션을 주면 기준열을 지정한다.

```python
sum(raw.duplicated())
```

```python
sum(raw.duplicated(subset=['도서명', '저자', 'ISBN']))
```

```python
# 중복된 행을 찾는 작업
# keep=False: 중복된 행을 모두 True로 표시한 불리언 배열을 반환하는 옵션
dup_row = raw.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
raw[dup_row].head()
```

#### 그룹별로 데이터 추출

```python
# 중복된 것을 제거하고 한권으로 처리하는 작업
# groupby(by=기준열, dropna=True/False)
group_df = count_df.groupby(by=['도서명', '저자', 'ISBN', '권', '대출건수'], dropna = False).sum()
group_df.head(10)
```
